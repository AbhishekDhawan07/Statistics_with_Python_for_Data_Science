## Mean and Median Calculation using NumPy.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50, 60])
mean = np.mean(data)
median = np.median(data)

print("Mean of the Data", mean)
print("Median of the Data", median)

```

## Standard Deviation and Variance Calculation with NumPy.

```python
data = np.array([10, 20, 30, 40, 50, 60])

std_dev = np.std(data)
variance = np.var(data)

print("Standard Deviation of the Data", std_dev)
print("Variance of the data", variance)
```

## Generating Random Variables withon a Range and Perform Statistics on that.

```python
# Generate the 100 Random Numbers from a Normal Distribution with a Mean 0 and standard deviation as 1.

data = np.random.normal(0, 1, 100)

# Compute the Mean and Standard Deviation of the Generated Data.

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

print("Mean", mean)
print("Median", median)
print("Standard Deviation", std_dev)
```

## Performing the Hypothesis Test.

```python
from scipy.stats import ttest_ind

# Generate Two Sets of Random Data.

data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1, 100)

# Perform a two - sample test.
t_stat, p_value = ttest_ind(data1, data2)

print("T - Statistic:", t_stat)
print("p - value", p_value)
```

## Descriptive Statistics using Pandas.

```python
import pandas as pd

# Create a DataFrame with the Sample Data.
data = {"A": [1, 2, 3, 4],
        "B": [10, 20, 30, 40],
        "C": [100, 200, 300, 400]}
df = pd.DataFrame(data)
print(df)



# Calculate Mean, Median and Standard Deviation of the data.


print("Mean of the Data:")
print(df.mean())


print("Median of the Data:")
print(df.median())

print("Standard Deviation of the Data:")
print(df.std())
```

## Calculation of the Correlation and Covariance of the Data.

```python
import pandas as pd

# Create a DataFrame with the Sample Data.
data = {"A": [1, 2, 3, 4],
        "B": [10, 20, 30, 40],
        "C": [100, 200, 300, 400]}
df = pd.DataFrame(data)
print(df)


# Calculation of the Correlation and Covariance.
print("Correlation of the Data")
print(df.corr())


# Calculation of the Covariance.
print("Covariance of the Data")
print(df.cov())
```

## Groupby and Aggreation of the Data.

```python
data = {"Category" : ["A", "B", "C", "B", "A"],
        "Value"  : [10, 50, 20, 30, 40]}
df = pd.DataFrame(data)

# Groupby Category and Calculate the Mean.
grouped_df = df.groupby("Category")
mean_values = grouped_df.mean()

print(mean_values)
```

## Descriptive Statistics using SciPy.

```python
import numpy as np
from scipy import stats

data = np.array([1, 2, 3, 4])

# Compute Mean.
mean = np.mean(data)
print("Mean of the Data", mean)

# Compute Median.
median = np.median(data)
print("Median of the Data", median)

# Compute Standard Deviation.
std_dev = np.std(data)
print("Standard Deviation of the Data", std_dev)

# Compute Variance.
variance = np.var(data)
print("Variance of the Data", variance)
```

## Hyothesis Testing using SciPy.

```python
from scipy.stats import norm

# Create a Normal distribution with Mean = 0 and Standard Deviation = 1.
distribution = norm(0, 1)

# Compute the Probability Density Function i.e. PDF at a given value.
pdf_value = distribution.pdf(0.5)

# Compute the Cummlative Density Function i.e. CDF at a given value.
cdf_value = distribution.cdf(0.5)

print("PDF of the data", pdf_value)
print("CDF of the data", cdf_value)
```
