```python
import numpy as np
arr = np.array([2, 3, 4, 6, 7, 8, 9, 12, 13, 16, 17, 23, 25, 27, 34, 37, 201])
print(arr)
```

```python
Q1 = np.percentile(arr, 25)
Q3 = np.percentile(arr, 75)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
Upper_Bound = Q3 + (1.5*IQR)
print(Upper_Bound)
Lower_Bound = Q1 - (1.5*IQR) 
print(Lower_Bound)
```

```python
l = []
for i in arr:
    if i >=Lower_Bound and i  <=Upper_Bound:
        l.append(i)
arr2 = np.array(l)
print(arr)
print(arr2)

# values above or equal to Lower_Bound and values less than or equal to Lower_Bound are only considered, all other values other than them are considered as Outliers.
# So, here 201 as a value is considered as a Outlier and removed.
```

```python
import seaborn as sns 
sns.boxplot(x = arr) # With outlier
```

```python
sns.boxplot(arr2) # Without Outliers.
```
